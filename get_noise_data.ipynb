```python
import os
from IPython.display import FileLink, FileLinks
import pandas as pd
```

```python
file_path = "106-5.2.csv"
df = pd.read_csv('/media/tuan/NGUI GA/ECG_cleaned/{}'.format(file_path))
```

```python
df
```

```python
import random
```

```python
rand_list = list(range(0, df.shape[0] // 38400))
random.shuffle(rand_list)
```

```python
list1 = rand_list[:len(rand_list) // 5]
```

```python
list2 = rand_list[len(rand_list) // 5 : 2 * len(rand_list) // 5]
```

```python
list3 = rand_list[2 * len(rand_list) // 5: 3 * len(rand_list) // 5]
```

```python
list4 = rand_list[3 * len(rand_list) // 5: 4 * len(rand_list) // 5] 
```

```python
list5 = rand_list[4 * len(rand_list) // 5:]
```

```python
DATA_PER_5_MINUTE = 38400
current_file_index = 0
for i in range(0, df.shape[0], DATA_PER_5_MINUTE):
    frame_start_index = i
    if frame_start_index < 0:
        frame_start_index = 0
    frame_end_index = frame_start_index + DATA_PER_5_MINUTE
    if frame_end_index > df.shape[0]:
        break
    split_file_name = "{}.csv".format(current_file_index)
    
    j = i // 38400
    if (j in list1):
        SAVE_FILE_DIR = './106-5.2/1/'
    if (j in list2):
        SAVE_FILE_DIR = './106-5.2/2/'
    if (j in list3):
        SAVE_FILE_DIR = './106-5.2/3/'
    if (j in list4):
        SAVE_FILE_DIR = './106-5.2/4/'
    if (j in list5):
        SAVE_FILE_DIR = './106-5.2/5/'        

    split_csv = df['# ECG Channel 1'][frame_start_index:frame_end_index]
    save_file_path = os.path.join(SAVE_FILE_DIR, split_file_name)

    split_csv.to_csv(save_file_path, index=False)
    current_file_index += 1
```
