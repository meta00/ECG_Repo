```python
import os
import pandas as pd
```

```python
file_path = "096-1.csv"
df = pd.read_csv('D:\\ECG_cleaned\\{}'.format(file_path))
```

```python
#Read mapping file
mapping_df = pd.read_csv('mapping.csv')
```

```python
df
```

```python
SAMPLING_RATE = 256
DATA_PER_MINUTE = SAMPLING_RATE * 60
DATA_PER_5_MINUTE = DATA_PER_MINUTE * 5

OVERLAPPED_RATIO = 0.15
```

```python
SAVE_FILE_DIR = "D:\\ECG_split\\"
DATA_PATH = "D:\\ECG_cleaned\\"


current_file_index = 1
ground_truth_df = pd.DataFrame(columns = ['usubjid', 'timepoint', 'class'
    , 'ablett_score', 'file', 'original', 'length', 'start', 'stop'])

for r, d, f in os.walk(DATA_PATH):
        for file in f:
            if file.endswith(".csv"):
                print(file)
                for i in range(0, df.shape[0], DATA_PER_5_MINUTE):
                    frame_start_index = i - int(DATA_PER_5_MINUTE * OVERLAPPED_RATIO)
                    if frame_start_index < 0:
                        frame_start_index = 0
                    frame_end_index = frame_start_index + DATA_PER_5_MINUTE
                    if frame_end_index > df.shape[0]:
                        frame_end_index = df.shape[0]
                    file_info = mapping_df[mapping_df['file'] == file[:-4]]
                    file_length = frame_end_index - frame_start_index
                    #print(file_info)
                    #print(file_info['usubjid'].values[0])

                    split_file_name = "{}.csv".format(current_file_index)

                    split_csv = df['# ECG Channel 1'][frame_start_index:frame_end_index]
                    save_file_path = os.path.join(SAVE_FILE_DIR, split_file_name)

                    split_csv.to_csv(save_file_path, index=False)

                    file_row = {'usubjid' : file_info['usubjid'].values[0],
                                'timepoint' : file_info['timepoint'].values[0],
                                'class' : file_info['class'].values[0],
                                'ablett_score' : file_info['ablett_score'].values[0],
                                'file' : split_file_name,
                                'original': file_info['file'].values[0],
                                'length': file_length,
                                'start': frame_start_index,
                                'stop': frame_end_index}

                    current_file_index += 1
                    ground_truth_df = ground_truth_df.append(file_row, ignore_index=True)
                ground_truth_df.to_csv('gt.csv', index=False)
```
